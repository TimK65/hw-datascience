```python
print(time.strftime("%H:%M:%S"))

# Fit a more complex linear regression model

features = ['Length', 'Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak']
label = 'AVG_TRIPS'

lm.fit(merged_data_dums.loc[:, features], merged_data_dums[label])

print(time.strftime("%H:%M:%S"))
```

# Random Forest

```python
# Well, this sucks. Let's try some random forest modeling.

print(time.strftime("%H:%M:%S"))

X = merged_data_dums[['Length', 'Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak', 'AVG_TRIPS']].dropna()
# X = merged_data_dums[['Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak', 'AVG_TRIPS']].dropna()
y = X['AVG_TRIPS']
X.drop('AVG_TRIPS', axis=1, inplace=True)

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 20)
model.fit(X, y)

print(time.strftime("%H:%M:%S"))
```

```python
features = X.columns
feature_importances = model.feature_importances_

features_df = pd.DataFrame({'Features': features, 'Importance Score': feature_importances})
features_df.sort_values('Importance Score', inplace=True, ascending=False)

features_df.head()
```

```python
from sklearn.cross_validation import cross_val_score
print(model)
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error')
print('CV MSE {}, Average MSE {}'.format(scores, scores.mean()))

auc_scores = {}
for n_trees in range(1, 100, 10):
    model = RandomForestRegressor(n_estimators = n_trees)
    scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error')
    print('n trees: {}, CV MSE {}, Average MSE {}'.format(n_trees, scores, scores.mean()))
    auc_scores[n_trees] = scores.mean()
pd.Series(auc_scores, name='MSE Scores').plot();
```

# Simple Decision Tree

```python
# CRASH!!! Um, okay. Let's try some simple decision-tree modeling.

from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz # to visualize decision tree using graphviz http://www.graphviz.org/
from sklearn.externals.six import StringIO
import pydotplus as pydot # python interface to graphviz dot language https://github.com/erocarrera/pydot
from os import system
from IPython.display import Image # to display graphviz image in notebook

X = merged_data_dums[['Length', 'Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak', 'AVG_TRIPS']].dropna()
# X = merged_data_dums[['Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak', 'AVG_TRIPS']].dropna()
y = X['AVG_TRIPS']
X.drop('AVG_TRIPS', axis=1, inplace=True)

# Instantiating our model object
model = DecisionTreeRegressor(max_depth=5) # Setting depth=5 (to ask only five questions)

# Fits the model
model.fit(X, y)

def show_tree(model):
    """Displays graphviz figure in notebook"""
    dot_data = StringIO()
    export_graphviz(model, 
                    out_file = dot_data, 
                    feature_names = X.columns, 
                    filled=True, 
                    rounded=True)
    graph = pydot.graph_from_dot_data(dot_data.getvalue())
    return Image(graph[0].create_png())
#    return Image(graph.create_png())


def save_tree_image(model):
    """Saves graphviz figure in working directory"""    
    dotfile = open("tree.dot", 'w')
    export_graphviz(model,
                    out_file = dotfile,
                    feature_names = X.columns)

    dotfile.close()
    system("dot -Tpng tree.dot -o tree.png")
    
# show_tree(model)
save_tree_image(model)
```

```python
for depth in range(2, 30):
    model = DecisionTreeRegressor(max_depth=depth)
    scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
#     print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
    print(depth, scores.mean())
```

```python
# So depth 4 seems to be the (relatively) sweet spot.

# Instantiating our model object
model = DecisionTreeRegressor(max_depth=4) # Setting depth=4 (to ask only four questions)

# Fits the model
model.fit(X, y)
```

```python
# But this MSE is still crappy, so let's try some other combinations of features to see if we can do better.

print(time.strftime("%H:%M:%S"))

# The random forest modeling above identified Length and Transfer as the most important features.

X = merged_data_dums[['Length', 'Transfer', 'AVG_TRIPS']].dropna()
# X = merged_data_dums[['Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak', 'AVG_TRIPS']].dropna()
y = X['AVG_TRIPS']
X.drop('AVG_TRIPS', axis=1, inplace=True)

for depth in range(2, 30):
    model = DecisionTreeRegressor(max_depth=depth)
    scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
#     print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
    print(depth, scores.mean())

print(time.strftime("%H:%M:%S"))
```

```python
print(time.strftime("%H:%M:%S"))

from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, export_graphviz # to visualize decision tree using graphviz http://www.graphviz.org/
from sklearn.externals.six import StringIO
import pydotplus as pydot # python interface to graphviz dot language https://github.com/erocarrera/pydot
from os import system
from IPython.display import Image # to display graphviz image in notebook
from sklearn.cross_validation import cross_val_score

X = merged_data_dums[['Length', 'Transfer', 'AVG_TRIPS']].dropna()
# X = merged_data_dums[['Transfer', 'daypart_AM Peak', 'daypart_Evening', 'daypart_Late Night Peak', 'daypart_Midday', 'daypart_PM Peak', 'AVG_TRIPS']].dropna()
y = X['AVG_TRIPS']
X.drop('AVG_TRIPS', axis=1, inplace=True)

model = DecisionTreeRegressor(
                max_depth = 10,
                min_samples_leaf = 2000)

model.fit(X, y)
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
#     print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
print(scores.mean())

# show_tree(model)
features = X.columns
feature_importances = model.feature_importances_

features_df = pd.DataFrame({'Features': features, 'Importance Score': feature_importances})
features_df.sort_values('Importance Score', inplace=True, ascending=False)

print(features_df.head())

print(time.strftime("%H:%M:%S"))
```
